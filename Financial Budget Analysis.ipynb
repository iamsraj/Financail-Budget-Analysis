```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('India_budget_2021.csv')
data
```

```python
data.shape
```

```python
#checked the missing value
data.isna().sum()
```

```python
#dropped the missing value
data=data.dropna()
```

```python
data.isna().sum()
```

### Ministry with highest Budget Allocation

```python
data.plot.bar(x = 'Department /Ministry', y = 'Fund allotted(in â‚¹crores)', figsize = (14,10))
```

```python
#Dropped the grand total
data=data.iloc[:-1 , :]
```

```python
data.plot.bar(x = 'Department /Ministry', y = 'Fund allotted(in â‚¹crores)', figsize = (14,10))
```

**So, now we can see Ministry of Finance is the one with the highest budget been allocated**

### Budget allocated to top 10 Ministries

```python
data_sorted  = data.sort_values('Fund allotted(in â‚¹crores)',ascending=False)
data_sorted.head(10)
```

```python
new_data = data_sorted.head(10)
new_data.plot.bar(x='Department /Ministry', y='Fund allotted(in â‚¹crores)',figsize=(12,6))  
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show
```

```python
df = new_data["Fund allotted(in â‚¹crores)"]
labels = new_data["Department /Ministry"]
plt.figure(figsize=(10,9))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=45,pctdistance=0.7,  shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show()
```

**We can see Ministry of Fianance is one with highest budget allocated to follwed by Ministry of Defence and so on.**
